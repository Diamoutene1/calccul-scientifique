### Mesures de dispersion des températures.

Nous souhaitons désormais mesurer la dispersion des températures, c'est-à-dire regarder comment les valeurs se répartissent autour de la moyenne. 
Les valeurs min et max ne donnent pas une bonne indication de la dispersion. 
La variance et l'écart type sont liés à la manière dont les points diffèrent de la moyenne. Plus la variance est grande, plus la dispersion est forte. 
La variance se calcule avec la formule : $$variance = \frac{\sum_i(x_i-x_{moy})^2}{n}$$ 
L'écart-type est la racine carré de la variance : $$ ect = \sqrt{variance}$$
