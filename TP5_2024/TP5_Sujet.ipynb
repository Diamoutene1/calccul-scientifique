- Tracer, sur un même graphique, le graphe des fonctions $\cos(x)$ et $\cos^2(x)$, pour 100 valeurs de $x$ comprises entre -5 et 5. 

- Ajouter les légendes et les propriétés de style de manière à ce que le résultat soit conforme à ce qui vous est donné ci-dessous. Il est obligatoire d'utiliser numpy et matplotlib.

```python
# A Faire ...

x=np.linspace(-5,5,100)
y1=np.cos(x)
y2=np.cos(x)**2
courb=plt.plot(x,y1,label="cos(x)")
courb1=plt.plot(x,y2,label="cos(x)**2")
plt.xlabel("abscisses")
plt.ylabel("ordonnees")
```

- Tracer les mêmes courbes, mais cette fois-ci chacune sur un shéma sur deux zones côte à côte comme sur le shéma ci-dessous:

```python
fig, ax = plt.subplots(ncols=2, nrows=1,figsize=(12, 2))

# A Compléter ...
ax[0].plot(x,y1,label="cos(x)")
ax[1].plot(x,y2,label="cos(x)**2")
```
