# Exercice 5 : logique


On veut travailler sur la formule F définie par:

$F= ( p \vee \neg (q \vee r)) \to \neg \Big( (\neg r \to q) \wedge \neg (p \vee q)\Big)$
    

Définir F avec sympy.

```python
import sympy as sp 
p,q,r=sp.symbols('p,q,r')
# A compléter ......
F=(p|~(q|r))>>~((~r>>q)&~(p|q))
display(F)
```

Que vaut F si on a p=True, q=False et r=False?

```python
# A compléter ......
premier=F.subs({p:True,q:False,r:False})
print(premier)

```

Déterminer une formule équivalente à F écrite sous forme Conjonctive.

```python
# A compléter ......
fnc=sp.to_cnf(F)
print(fnc)
```

Faire une table de vérité pour la formule F comme ci-dessous.

```python
# A compléter ......
for (a,b,c) in [(i,j,k)for i in(True,False) for j in (True,False) for k in(True,False)]:
    print(a,b,c,F.subs({p:a,q:b,r:c}))
```
